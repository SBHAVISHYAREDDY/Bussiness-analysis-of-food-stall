Kedepannya bisa dioptimalkan AOV dengan membuat promo paket bundling dengan harga yg lebih terjangkau (upselling)

```python
df_totalbill = df_AB_fill.groupby('Order_ID', as_index=False)['SubTotal'].sum()
df_totalbill.rename(columns={'SubTotal':'TotalBill'}, inplace=True)
```

```python
range_bill = []

for i, j in df_totalbill.iterrows(): 
    if j['TotalBill'] < 50000:
        bil = '<50K'
    elif j['TotalBill'] >= 50000 and j['TotalBill'] <= 100000:
        bil = '50K-100K'
    elif j['TotalBill'] > 100000 and j['TotalBill'] <= 200000:
        bil = '100K-200K'
    else:
        bil = '>200K'
    range_bill.append(bil)
```

```python
df_totalbill['RangeBill'] = range_bill
df_rangebill = pd.DataFrame(df_totalbill['RangeBill'].value_counts().reset_index().values, columns=["RangeBill", "Total Amount"])
df_rangebill
```

Membuat Variable baru untuk menghitung total transaksi berdasarkan Range Total Bill customer

```python
ax4 = sns.barplot(x='RangeBill', y='Total Amount', data=df_rangebill)
for i in ax4.containers:
    ax4.bar_label(i,)
```
